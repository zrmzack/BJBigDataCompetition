```python
import numpy as np
import pandas as pd
```

```python
with pd.ExcelFile('赛题三：疫情防控重点人群预测与画像-样例数据.xlsx') as excel:
    df1=pd.read_excel(excel,0)
    df2=pd.read_excel(excel,1)
    df3=pd.read_excel(excel,2)
    df4=pd.read_excel(excel,3)
    df5=pd.read_excel(excel,4)
```

```python
df12=pd.merge(df1,df2,on='用户标识',how='outer')
```

```python
df12['业务离开时间'].dtypes
```

```python
df123=pd.merge(df12,df3,on=['LAC','CI'],how='outer')
```

```python
df123.head()
```

```python
df1234=pd.merge(df123,df4,on=['LAC','CI'],how='outer',suffixes=('_基站对象','_北京中高风险地区'))
```

```python
df1234.head()
```

```python
df12345=pd.merge(df1234,df5,on='用户标识',how='outer')
```

```python
df12345.head()
```

```python
df12345.to_csv('合并-疫情防控重点人群预测与画像-样例数据.csv')
```
